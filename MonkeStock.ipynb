```python
import pandas as pd
def load_data():
  url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
  html = pd.read_html(url, header = 0)
  df = html[0]
  return df

```

```python
stock = 'MMM'
```

```python
df = load_data()
df
```

```python
import yfinance as yf
```

```python
list(df.Symbol)
```

```python
spy = {'Symbol': '^GSPC'}
df = df.append(spy, ignore_index=True)
```

```python
data = yf.download(
    tickers = list(df.Symbol),
    period = "1y",
    interval = "1d",
    group_by = 'ticker',
    auto_adjust = True,
    threads = True,
    proxy = None
)
```

```python
data
```

```python
data[stock]
```

```python
df = pd.DataFrame(data[stock].Close)
df['Date'] = df.index
df
```

```python
df = pd.DataFrame(data[stock].Close.pct_change() * 100)
df['Date'] = df.index
df
```

```python
import matplotlib.pyplot as plt
```

```python
def plotPerformance(arg):
    df = arg
    
    plt.figure(figsize=(16,9))

    #  subplot #2
    plt.subplot(222)
    plt.plot(df.index, data[stock].Close)
    plt.plot(df.index, data['AMZN'].Close)
    plt.plot(df.index, data['AAPL'].Close)
    plt.plot(df.index, data['ALGN'].Close)
    plt.title("stock")
    #plt.legend(df['AAPL'], loc="upper left", bbox_to_anchor=(1,1))
    plt.xticks(rotation = 90)
    
    plt.show()  # plot subplots
```
