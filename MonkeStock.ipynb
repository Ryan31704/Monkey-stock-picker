```python
import math
import pandas as pd
def load_data():
  url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
  html = pd.read_html(url, header = 0)
  df = html[0]
  return df

```

```python
spy = {'Symbol': '^GSPC'}
df_spy = df_spy.append(spy, ignore_index=True)
```

```python
data = yf.download(
    tickers=list(df_spy.Symbol),
    period = "1y",
    interval = "1d",
    group_by = 'ticker',
    auto_adjust = True,
    threads = True,
    proxy = None
)
```

```python
import random

df_spy = load_data()
symbol_list = list(df_spy.Symbol)
symbol1 = random.choice(symbol_list)
symbol2 = random.choice(symbol_list)
symbol3 = random.choice(symbol_list)
```

```python
import matplotlib.pyplot as plt
from cProfile import label


def plotPerformance(arg):
    df = arg
    
    plt.figure(figsize=(16,9))

    #  subplot #2
    plt.subplot(222)
    plt.plot(df.index, data[symbol1].Close, label="".join(symbol1))
    plt.plot(df.index, data[symbol2].Close, label="".join(symbol2))
    plt.plot(df.index, data[symbol3].Close, label="".join(symbol3))
    plt.title("Random Stocks")
    plt.legend(loc="upper right")
    plt.xticks(rotation = 90)
    plt.show()  # plot subplots
```

```python
plotPerformance(data)
```
